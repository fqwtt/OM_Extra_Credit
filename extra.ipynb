```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
Nasdaq_100 = pd.read_excel('Nasdaq100companies.xlsx')
Nasdaq_100.head()
```

```python
Nasdaq_100.iloc[:,2]
```

```python
Nasdaq_100_list = Nasdaq_100.iloc[:,2].tolist()
Nasdaq_100_list
```

```python
hist = yf.download(Nasdaq_100_list, start="2021-01-01", end="2021-12-31", interval = '1d',group_by = 'ticker')
```

```python
hist_df = pd.DataFrame(hist)
hist_df.head()
```

```python
hist_df.columns = ['_'.join(col).strip() for col in hist_df.columns.values]
hist_df.head()
```

```python
hist_df_close = hist_df.filter(regex='Close')
hist_df_close.head()
```

```python
hist_df_close_return = hist_df_close.pct_change()
hist_df_close_return.head()
```

```python
hist_df_close_return.columns = [x.split('_')[0]+'_daily_return' for x in hist_df_close_return.columns]
hist_df_close_return.head()
```

```python
hist_df_close_return = hist_df_close_return.iloc[:,::2]
hist_df_close_return.head()
```

```python
hist_df_close_return.to_csv('Nasdaq_100_Daily.csv')
```

```python
corr_matrix = hist_df_close_return.corr()
corr_matrix
```

```python
corr_matrix.to_csv('Nasdaq_100_Daily_Corr.csv')
```

```python
from pyomo.environ import *

# define input parameters
Company = Nasdaq_100.iloc[:,2].tolist()
Return = hist_df_close_return.values.tolist()

model = ConcreteModel()

# define decision variables (i.e. allocation to bonds in percentage)
model.x = Var(Company, within=NonNegativeReals)

# define objective function
model.obj = Objective(
    expr = sum(Return[i]*model.x[i] for i in Company),
    sense = maximize )

# total percentage sum to 1
model.onehundred_con = Constraint(
    expr = sum(model.x[i] for i in Company)==1   )

# absolute value of the difference between our portfolio and the nasdaq 100
model.nasdaq_con = Constraint(
    expr = sum(Return[i]*model.x[i] for i in Company) - sum(Return[i] for i in Company) == 0   )

# solve the model
opt = SolverFactory('glpk')
opt_success = opt.solve(model)

# print the solutions
print('Portfolio Return:', value(model.obj))

print( 'Company', 'Allocation')
print('=========================')
for i in Company:
    print(i, ':',  value(model.x[i]))

```
