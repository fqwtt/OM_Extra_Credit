```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
Nasdaq_100 = pd.read_excel('Nasdaq100companies.xlsx')
Nasdaq_100.head()
```

```python
Nasdaq_100.iloc[:,2]
```

```python
Nasdaq_100_list = Nasdaq_100.iloc[:,2].tolist()
Nasdaq_100_list
```

```python
hist = yf.download(Nasdaq_100_list, start="2021-01-01", end="2021-12-31", interval = '1d',group_by = 'ticker')
```

```python
hist_df = pd.DataFrame(hist)
hist_df.head()
```

```python
hist_df.columns = ['_'.join(col).strip() for col in hist_df.columns.values]
hist_df.head()
```

```python
hist_df_close = hist_df.filter(regex='Close')
hist_df_close.head()
```

```python
hist_df_close_return = hist_df_close.pct_change()
hist_df_close_return.head()
```

```python
hist_df_close_return.columns = [x.split('_')[0]+'_daily_return' for x in hist_df_close_return.columns]
hist_df_close_return.head()
```

```python
hist_df_close_return = hist_df_close_return.iloc[:,::2]
hist_df_close_return.head()
```

```python
hist_df_close_return.to_csv('Nasdaq_100_Daily.csv')
```
